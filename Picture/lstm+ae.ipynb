```python
import  matplotlib.pyplot as plt
```

```python
accuracy = [0.90296, 0.91031, 0.91464, 0.99661, 0.99690]
```

```python
precision = [0.90298, 0.92531, 0.93486, 0.98818, 0.99052]
```

```python
recall = [0.95872, 0.95872, 0.97328, 0.97976, 0.99998]
```

```python
f1 = [0.94901, 0.95176, 0.953684, 0.97322, 0.97436]
```

```python
name = ['Combination one', 'Combination two', 'Combination three', 'Combination four', 'Combination five']
```

```python
x = list(range(len(accuracy)))
total_width, n = 0.4, 2
```

```python
width = total_width / n
```

```python
plt.figure(figsize = (10, 6))

plt.plot(name, accuracy,'r--',label='accuracy')
for a,b in zip(x, accuracy):   #柱子上的数字显示
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10)

plt.plot(name, precision,'g--',label='precision')
for a,b in zip(x, precision):   #柱子上的数字显示
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10)

plt.plot(name, recall,'b--',label='recall')
for a,b in zip(x, recall):   #柱子上的数字显示
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10)

plt.plot(name, f1,'y--',label='f1-score')
for a,b in zip(x, f1):   #柱子上的数字显示
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10)

plt.plot(name, accuracy,'ro-', name, precision,'g+-', name, recall,'b^-', name, f1,'y+')

plt.legend()
plt.savefig('./lstmae1.jpg')
plt.show()
```

```python
train = [1119, 1178, 1119, 758, 776]
```

```python
test = [32, 31, 30, 34, 27]
```

```python
time = [1580, 1697, 1699, 1347, 1298]
```

```python
plt.figure(figsize = (12, 8))
plt.bar(x, train, width = width, label = 'train_time', fc = 'orange')
for a,b in zip(x, train):   #柱子上的数字显示
 plt.text(a,b,'%.0f'%b,ha='center',va='bottom',fontsize=10);
for i in range(len(x)) :
    x[i] = x[i] + width
plt.bar(x, test, width = width, label = 'test_time', fc = 'b')
for a,b in zip(x, test):   #柱子上的数字显示
 plt.text(a,b,'%.0f'%b,ha='center',va='bottom',fontsize=10);
for i in range(len(x)) :
    x[i] = x[i] + width
plt.bar(x, time, width = width, label = 'alltime', tick_label = name, fc = 'g')
for a,b in zip(x, time):   #柱子上的数字显示
 plt.text(a,b,'%.0f'%b,ha='center',va='bottom',fontsize=10);
plt.legend()
plt.savefig('./ae2.jpg')
plt.show()
```

```python
auc = [0.906, 0.927, 0.925, 0.975, 0.998]
```

```python
plt.figure(figsize = (12, 6))
plt.bar(x, auc, width = width, label = 'AUC', tick_label = name, fc = 'g')
for a,b in zip(x, auc):   #柱子上的数字显示
 plt.text(a,b,'%.3f'%b,ha='center',va='bottom',fontsize=10);
plt.legend()
plt.savefig('./ae3.jpg')
plt.show()
```
