```python
x = list(range(len(accuracy)))
total_width, n = 0.4, 2
```

```python
width = total_width / n
```

```python
plt.figure(figsize = (16, 12))
plt.bar(x, accuracy, width = width, label = 'accuracy', fc = 'y')
for a,b in zip(x, accuracy):   #柱子上的数字显示
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10);
for i in range(len(x)) :
    x[i] = x[i] + width
plt.bar(x, precision, width = width, label = 'precisioon', fc = 'b')
for a,b in zip(x, precision):   #柱子上的数字显示
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10);
for i in range(len(x)) :
    x[i] = x[i] + width
plt.bar(x, recall, width = width, label = 'recall', fc = 'orange')
for a,b in zip(x, recall):   #柱子上的数字显示
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10);
for i in range(len(x)) :
    x[i] = x[i] + width
plt.bar(x, f1, width = width, label = 'f1-score', tick_label = name, fc = 'g')
for a,b in zip(x, f1):   #柱子上的数字显示
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10);
plt.legend()
plt.savefig('./ae1.jpg')
plt.show()
```
