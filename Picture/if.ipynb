```python
import  matplotlib.pyplot as plt
```

```python
accuracy = [0.81691, 0.20061, 0.20062, 0.71863, 0.71863]
```

```python
precision = [0.99556, 0.86347, 0.86347, 0.93400, 0.93400]
```

```python
recall = [0.77607, 0.00856, 0.00856, 0.70002, 0.70002]
```

```python
f1 = [0.87222, 0.01695, 0.01695, 0.80026, 0.80026]
```

```python
name = ['Combination one', 'Combination two', 'Combination three', 'Combination four', 'Combination five']
```

```python
x = list(range(len(accuracy)))
total_width, n = 0.4, 2
```

```python
width = total_width / n
```

```python
plt.figure(figsize = (10, 6))

plt.plot(name, accuracy,'r--',label='accuracy')
for a,b in zip(x, accuracy):   #
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10)

plt.plot(name, precision,'g--',label='precision')
for a,b in zip(x, precision):   #
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10)

plt.plot(name, recall,'b--',label='recall')
for a,b in zip(x, recall):   #
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10)

plt.plot(name, f1,'y--',label='f1-score')
for a,b in zip(x, f1):   #
 plt.text(a,b,'%.4f'%b,ha='center',va='bottom',fontsize=10)

plt.plot(name, accuracy,'ro-', name, precision,'g+-', name, recall,'b^-', name, f1,'y+')

plt.legend()
plt.savefig('./if1.jpg')
plt.show()
```

```python
train = [2, 0.8, 0.7, 0.6, 0.5]
```

```python
test = [15, 11, 12, 10, 10]
```

```python
time = [18, 15, 16, 14, 14]
```

```python
plt.figure(figsize = (12, 8))
plt.bar(x, train, width = width, label = 'train_time', fc = 'orange')
for a,b in zip(x, train):   #
 plt.text(a,b,'%.0f'%b,ha='center',va='bottom',fontsize=10);
for i in range(len(x)) :
    x[i] = x[i] + width
plt.bar(x, test, width = width, label = 'test_time', fc = 'b')
for a,b in zip(x, test):   #
 plt.text(a,b,'%.0f'%b,ha='center',va='bottom',fontsize=10);
for i in range(len(x)) :
    x[i] = x[i] + width
plt.bar(x, time, width = width, label = 'alltime', tick_label = name, fc = 'g')
for a,b in zip(x, time):   #
 plt.text(a,b,'%.0f'%b,ha='center',va='bottom',fontsize=10);
plt.legend()
plt.savefig('./if2.jpg')
plt.show()
```

```python
auc = [0.881, 0.501, 0.501, 0.745, 0.748]
```

```python
plt.figure(figsize = (12, 6))
plt.bar(x, auc, width = width, label = 'AUC', tick_label = name, fc = 'g')
for a,b in zip(x, auc):   #
 plt.text(a,b,'%.3f'%b,ha='center',va='bottom',fontsize=10);
plt.legend()
plt.savefig('./if3.jpg')
plt.show()
```
